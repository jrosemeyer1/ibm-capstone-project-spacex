```python
# Task 2: Mark the success/failed launches for each site on the map

# --- Helper: show Folium map inline even in restricted environments ---
from IPython.display import HTML, display
from urllib.parse import quote

def show_folium_inline(m, width=800, height=500):
    """
    Renders a Folium map inline by embedding the full HTML into a data URI iframe.
    Works even when the notebook isn't trusted or if IFrame('file.html') is blocked.
    """
    html = m.get_root().render()
    data_uri = "data:text/html;charset=utf-8," + quote(html)
    return HTML(f'<iframe src="{data_uri}" width="{width}" height="{height}" style="border:none;"></iframe>')


# --- TASK 2: Mark the success/failed launches for each site on the map (inline, no saving) ---
import folium
from folium.plugins import MarkerCluster

# Assumes spacex_df exists with columns: ['Launch Site','Lat','Long','class']
# If not, ensure you've run the earlier cells that load and prepare spacex_df.

# Center map on mean coordinates for a nice initial view
mean_lat = spacex_df['Lat'].mean()
mean_long = spacex_df['Long'].mean()
spacex_map = folium.Map(location=[mean_lat, mean_long], zoom_start=5)

# Cluster nearby points
marker_cluster = MarkerCluster().add_to(spacex_map)

# Add markers (green = success, red = failure)
for _, record in spacex_df.iterrows():
    lat = record['Lat']
    lon = record['Long']
    site = record['Launch Site']
    success = int(record['class'])  # 1 or 0
    color = 'green' if success == 1 else 'red'

    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=f"Site: {site}<br>Outcome: {'Success' if success == 1 else 'Failure'}"
    ).add_to(marker_cluster)

# INLINE display (no file write; executes even in limited environments)
display(show_folium_inline(spacex_map, width=900, height=560))

```

```python
# TASK: Add a folium.Marker for each launch to the marker cluster

# Reuse the inline Folium display function if not already defined
from IPython.display import HTML, display
from urllib.parse import quote

def show_folium_inline(m, width=800, height=500):
    html = m.get_root().render()
    data_uri = "data:text/html;charset=utf-8," + quote(html)
    return HTML(f'<iframe src="{data_uri}" width="{width}" height="{height}" style="border:none;"></iframe>')

# Create a fresh base map centered near NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Iterate through each launch and add a marker
for index, record in spacex_df.iterrows():
    lat = record['Lat']
    lon = record['Long']
    site = record['Launch Site']
    outcome = 'Success' if record['class'] == 1 else 'Failure'
    color = record['marker_color']

    # Create and add a marker
    marker = folium.Marker(
        location=[lat, lon],
        icon=folium.Icon(color='white', icon_color=color, icon='rocket', prefix='fa'),
        popup=f"Site: {site}<br>Outcome: {outcome}"
    )
    marker.add_to(marker_cluster)

# Add the cluster layer to the map
site_map.add_child(marker_cluster)

# Display the map inline (no saving or trusting required)
display(show_folium_inline(site_map, width=900, height=560))
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )

# --- Step 1: Setup MousePosition to display coordinates on hover ---
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# --- Step 2: Define function to calculate distance ---
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # approximate radius of Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon, dlat = lon2 - lon1, lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# --- Step 3: Coordinates for launch site & coastline ---
launch_site_lat, launch_site_lon = 28.563197, -80.576820   # CCAFS SLC-40
coastline_lat, coastline_lon = 28.56367, -80.57163         # coastline from map hover

# --- Step 4: Calculate the distance ---
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# --- Step 5: Add a marker for the coastline point with distance label ---
coastline_coordinate = [coastline_lat, coastline_lon]
distance_marker = folium.Marker(
    coastline_coordinate,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_coastline),
    ),
)
distance_marker.add_to(site_map)

# --- Step 6: Draw a line between the launch site and the coastline ---
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
    color='blue',
    weight=2,
).add_to(site_map)

# --- Step 7: Display map inline ---
display(show_folium_inline(site_map, width=900, height=560))
```
